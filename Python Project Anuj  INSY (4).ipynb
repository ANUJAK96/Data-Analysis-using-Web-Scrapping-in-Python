```python
# importing the libraries which are necessary for the work
from bs4 import BeautifulSoup#from bs4 we are importing beautifulsoup
import requests#we are importing request which will be using it to request the url of sites
import parser   #we are importing parser to reverse html
import csv     #we are importing csv to convert and write onto file
import os     #we are importing os to interact with the operating systems
inp=0 #initializing the null character to inp
# Step 1: Sending a HTTP request to a URL
link1 = "https://money.cnn.com/data/hotstocks/"           #initilizing the link1
# we are giving a GET request to bring the raw HTML of the overall url
html_content = requests.get(link1).text                 
link2 = "https://finance.yahoo.com/quote/"          # initilizing the link 2 
print("Which Stock are you interested in:")  #asking the users for choices
print("\nMost Actives:")
# We are  Parsing the html content
soup = BeautifulSoup(html_content, "lxml")
global str
if os.path.exists("stocks.csv"):
    os.remove("stocks.csv")     
# next step we Analyze the HTML tag, where the content lives
#  We Create a data dictionary that can store the data .
data = {}
count_one=0 #we are inititilizing counter to zero
count_two=1  # we are initilizing another counter to one
main_category ="Most Actives"  #for the main catogeries most actives
org_name =""  #input variable for company name
open_value =""  # input variable for open value
previous_close=""  #input variable for previous close
volume=""         #input variable for volume
market_cap =""    #input variable for market cap
ignore=True       
with open('stocks.csv', 'w', newline='') as file:    #we are opening the file in write mode 
    writer = csv.writer(file)                 #we are writing a csv file
    writer.writerow(["main_category", "ticker", "name","open","previous_close","volume","market_cap"]) #writing all the names of headers
    for company in soup.find_all("a", attrs={"class": "wsod_symbol"}):   #we are looping to get all the tag with 'a' and class with "wsod_symbol"
        if count_one>=3:#if the count is > 3
            link_cont2 = requests.get(link2+company.text).text  #the choice is added to the yahoo finance link to get those values
            soup2 = BeautifulSoup(link_cont2,"lxml")        #scrapping the link 2 to lxml
            for companyFullName in soup2.find_all("h1", attrs={"class": "D(ib)"}):    #we are getting all the details to tag 'h1' and 'class D(ib)'
                org_name = companyFullName.text #getting the companyname
                s= org_name.split(" (")   #converting to lists and getting the actual ticker value
                org_name =s[0]   #taking the 0th array of index which is company name
            for open_value in soup2.find("td", attrs={"data-test":"OPEN-value"}):#we are getting all the details from the tag 'td' and from datatest of class "OPEN-value"
                open_value = open_value.text     #converting into text format
            for closevalue in soup2.find("td", attrs={"data-test":"PREV_CLOSE-value"}):#getting all the details of close value from the tag "td" and from the datatest of class "PREV_CLOSE-value"
                previous_close = closevalue.text #converting to text format
            for volumevalue in soup2.find("td", attrs={"data-test":"TD_VOLUME-value"}):#getting all the details of  volume from the tag "td" and from the datatest of class "TD_VOLUME-value"
                volume = volumevalue.text               #converting into text format
                volume= volume.replace(",","")  #removing comma
            for market_cap in soup2.find("td", attrs={"data-test":"MARKET_CAP-value"}):#getting all the details of  market cap from the tag "td" and from the datatest of class "MARKET_CAP-value"
                market_cap = market_cap.text   #converting into text
                print(company.text+" "+org_name) #printing the full company name
                writer.writerow([main_category, company.text, org_name,open_value,previous_close,volume,market_cap])#writing all the values in a row in the file
            count_two=count_two+1#iterating it to next element
        count_one=count_one+1  #iterating it the overall
        if count_two==11: #From 11th element, it is gainers in the file
            main_category = "Gainers"
            print("\n\n"+main_category+":")
        elif count_two==21: #From 21st element, it is loosers in the file. 
            main_category = "loosers"
            print("\n\n"+main_category+":")
while inp==0:             #using while true statement
    choice = input("\n\nEnter your choice : ") #asking the choice from the user
    foundChoice = False  #creating a object, initializiing bool false
    with open('stocks.csv', 'r') as file:             #opening file in read mide
        csv_file_reader = csv.DictReader(file)      
        for row in csv_file_reader:                  # iterating all the rows in file
            if(row['ticker']==choice):               # checking for the company name
                foundChoice = True #goes inside only the choice is correct ie in the file
                print("\nThe data for "+choice+" is the following:")   #printing the choice and the values present in the row
                print(row['ticker']+" "+row['name'])  #printing whole name of the company
                print("OPEN: "+row['open'])          #printing the open value for the company
                print("PREV CLOSE: "+row['previous_close'])#printing the previous close value of the company
                print("VOLUME: "+f"{int(row['volume']):,}")#printing the volume of the company
                print("MARKET CAP: "+row['market_cap'])   #printing the market cap for the company
                q=input("Do you want to continue(y/n)") # asking user to continue or exit and moving the code as request
                if q=='n'or q=='N':     #do you want to continue statement 
                    inp=1
        if foundChoice==False:          #if choice is not correct ie not in the file
            print("\n\nPlease enter the right choice") #asking for right choice

            
```
